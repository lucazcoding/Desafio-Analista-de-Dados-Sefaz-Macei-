# Etapa 4 — Selecionar Estudos de Caso
## Melhor, Pior e Maceió

## Contexto

Nas etapas anteriores calculamos as taxas, comparamos as capitais e identificamos padrões nacionais. Agora vamos selecionar os três estudos de caso para aprofundamento na etapa 5.

## Objetivo

Selecionar três capitais para aprofundamento:

1. **Melhor capital** — maior Taxa Geral
2. **Pior capital** — menor Taxa Geral
3. **Maceió** — sempre, independente de onde estiver

## Perguntas que esta etapa responde

1. **"Quem é a melhor capital geral?"**
2. **"Quem é a pior capital geral?"**
3. **"Onde Maceió está no ranking geral?"**

## Visualização

**Nenhuma nova** — os casos emergem do scatter plot da etapa 3.

```python
import sys
from pathlib import Path
import pandas as pd

# Adiciona a raiz do projeto ao sys.path
sys.path.insert(0, str(Path.cwd().parent))
```

## 1. Carregar Dados da Etapa 1

```python
# Carregar dados
pasta_processados = Path.cwd().parent / 'data' / 'processed'

df_taxa_geral = pd.read_parquet(pasta_processados / 'etapa1_taxa_geral.parquet')

print(f'Taxa geral: {len(df_taxa_geral)} capitais')
df_taxa_geral.head()
```

## 2. Identificar os Três Estudos de Caso

```python
# Extrair nomes das capitais
df_taxa_geral['Capital'] = df_taxa_geral['Instituição'].apply(
    lambda x: x.replace('Prefeitura Municipal de ', '').replace('Prefeitura Municipal do ', '').split(' - ')[0].strip()
)

# Melhor capital
melhor = df_taxa_geral.sort_values('Taxa_Geral', ascending=False).iloc[0]

# Pior capital
pior = df_taxa_geral.sort_values('Taxa_Geral', ascending=True).iloc[0]

# Maceio
maceio = df_taxa_geral[df_taxa_geral['Capital'] == 'Maceió'].iloc[0]

print('=== ESTUDOS DE CASO SELECIONADOS ===')

print(f'\n1. MELHOR CAPITAL:')
print(f'   Capital: {melhor["Capital"]}')
print(f'   UF: {melhor["UF"]}')
print(f'   Taxa Geral: {melhor["Taxa_Geral"]:.2f}%')

print(f'\n2. PIOR CAPITAL:')
print(f'   Capital: {pior["Capital"]}')
print(f'   UF: {pior["UF"]}')
print(f'   Taxa Geral: {pior["Taxa_Geral"]:.2f}%')

print(f'\n3. MACEIÓ:')
print(f'   Capital: {maceio["Capital"]}')
print(f'   UF: {maceio["UF"]}')
print(f'   Taxa Geral: {maceio["Taxa_Geral"]:.2f}%')
```

## 3. Ranking de Maceió

```python
# Ranking de Maceio
df_ranking = df_taxa_geral.sort_values('Taxa_Geral', ascending=False).reset_index(drop=True)
df_ranking['Ranking'] = range(1, len(df_ranking) + 1)

posicao_maceio = df_ranking[df_ranking['Capital'] == 'Maceió']['Ranking'].values[0]
total_capitais = len(df_ranking)

print(f'\n=== POSICAO DE MACEIO ===')
print(f'Ranking: {posicao_maceio}º lugar de {total_capitais} capitais')
print(f'Taxa Geral: {maceio["Taxa_Geral"]:.2f}%')

# Comparar com media
media_geral = df_taxa_geral['Taxa_Geral'].mean()
if maceio['Taxa_Geral'] > media_geral:
    print(f'\nMaceio esta ACIMA da media nacional ({media_geral:.2f}%)')
else:
    print(f'\nMaceio esta ABAIXO da media nacional ({media_geral:.2f}%)')
```

## 4. Comparar os Três Casos

```python
# Criar DataFrame comparativo
casos = pd.DataFrame([
    {'Caso': 'Melhor', 'Capital': melhor['Capital'], 'UF': melhor['UF'], 'Taxa_Geral': melhor['Taxa_Geral']},
    {'Caso': 'Maceió', 'Capital': maceio['Capital'], 'UF': maceio['UF'], 'Taxa_Geral': maceio['Taxa_Geral']},
    {'Caso': 'Pior', 'Capital': pior['Capital'], 'UF': pior['UF'], 'Taxa_Geral': pior['Taxa_Geral']}
])

print('=== COMPARACAO DOS TRES CASOS ===\n')
casos
```

```python
# Diferencas
diff_melhor_maceio = melhor['Taxa_Geral'] - maceio['Taxa_Geral']
diff_maceio_pior = maceio['Taxa_Geral'] - pior['Taxa_Geral']
diff_melhor_pior = melhor['Taxa_Geral'] - pior['Taxa_Geral']

print('=== DIFERENCAS ===')
print(f'\nMelhor vs Maceio: {diff_melhor_maceio:.2f} pp')
print(f'Maceio vs Pior: {diff_maceio_pior:.2f} pp')
print(f'Melhor vs Pior: {diff_melhor_pior:.2f} pp')
```

## 5. Salvar Casos Selecionados

```python
# Salvar casos
casos.to_parquet(pasta_processados / 'etapa4_casos_selecionados.parquet', index=False)

# Salvar DataFrame completo com ranking
df_ranking.to_parquet(pasta_processados / 'etapa4_ranking_geral.parquet', index=False)

print('AArquivos salvos:')
print(f'  - {pasta_processados / "etapa4_casos_selecionados.parquet"}')
print(f'  - {pasta_processados / "etapa4_ranking_geral.parquet"}')
```
